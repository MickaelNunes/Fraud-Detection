### Split the game 

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns='Class')
y = df['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2)
```

### Unbalanced problem = Random oversampling: SMOTE 

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(sampling_strategy=0.1, random_state=20)
```

```python
X_new, y_new = sm.fit_resample(X_train, y_train)
```

### Normalization 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_new, y_new)
X_test_scaled = scaler.transform(X_test, y_test)
```

```python
sns.boxplot(data=X_train_scaled)
plt.show
```

### Random forest classifier 

```python
from sklearn.metrics import precision_recall_curve, precision_score, recall_score, auc
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

#param_grid = {'n_estimators': np.arange(70, 130, 10),'max_depth': np.arange(8, 17, 1),}

rfc = RandomForestClassifier(n_estimators=120,max_depth=14)
#gsc = GridSearchCV(rfc, param_grid)
rfc.fit(X_train_scaled, y_new)

#gsc.best_params_

from sklearn.metrics import precision_score, recall_score, auc
precision , recall, threshold = precision_recall_curve(y_test, rfc.predict(X_test_scaled))
print(auc(recall, precision))
plt.plot(recall, precision)
```

```python
precision , recall, threshold = precision_recall_curve(y_test, rfc.predict_proba(X_test_scaled)[:,1])
print(auc(recall, precision))
plt.plot(recall, precision)

precision_t , recall_t, threshold_t = precision_recall_curve(y_new, rfc.predict_proba(X_train_scaled)[:,1])
print(auc(recall_t, precision_t))
```

```python
import sklearn.metrics as metrics
fpr, tpr, threshold = metrics.roc_curve(y_test, rfc.predict(X_test))
roc_auc = metrics.auc(fpr, tpr)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f')
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

### Perform project (Isolation forest method)

```python
X_train['Class'] = y_train
X_new = X_train[X_train['Class']==0]
X_new = X_train.drop(columns='Class')

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_new_scaled = scaler.fit_transform(X_train, y_train[y_train==0])
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor

IF = IsolationForest(n_estimators=100, contamination="auto", behaviour="new", random_state=35) 
LOF = LocalOutlierFactor(contamination="auto", novelty=True) 

IF.fit(X_new)
LOF.fit(X_new)

ypredIF = IF.predict(X_test)
ypredLOF = LOF.predict(X_test)

# Warning, those algorithm return -1 for anomaly and 1 for normal!!
ypredIF[ypredIF == 1] = 0
ypredIF[ypredIF == -1] = 1
ypredLOF[ypredLOF == 1] = 0
ypredLOF[ypredLOF == -1] = 1
```

```python
precision_recall_curve(y_test, IF.predict(X_test_scaled))
```

```python
precision , recall, threshold = precision_recall_curve(y_test, IF.predict(X_test_scaled))
print("Isolation Forest results:", auc(recall, precision))
plt.plot(recall, precision)
```

```python
precision , recall, threshold = precision_recall_curve(y_test, LOF.predict(X_test_scaled))
print("LOF results:", auc(recall, precision))
plt.plot(recall, precision)
```
